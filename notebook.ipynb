## Exploratory Data Analysis (EDA)

### Explore ...

- Melakukan **merge** air quality dataset setiap kota menggunakan function **concat()**

```python
#Merge Dataset 
airquality_df = pd.concat([aotizhongxin_df,
                           changping_df,
                           dingling_df,
                           dongsi_df,
                           guanyuan_df,
                           gucheng_df,
                           huairou_df,
                           nongzhanguan_df,
                           shunyi_df,
                           tiantan_df,
                           wanliu_df,
                           wanshouxigong_df], ignore_index=True)
airquality_df.head()
```

- Menghitung nilai rata-rata polutan di 12 kota tahun 2013-2017.
- Function **groupby()** digunakan dalam mengklasifikasikan kolom _station_ dan function **agg()** untuk klasifikasi kolom setiap jenis polutan.
- Perhitungan _mean_ dilakukan dengan menuliskan **"PM2.5": ["mean"]** untuk setiap jenis polutan dalam function agg().
- Function **round(2)** untuk menampilkan 2 angka setelah koma.

```python
#Mean Cluster of each Pollutants on Selected Stations
airquality_df.groupby(by= 'station').agg({
    "PM2.5": ["mean"],
    "PM10" : ["mean"],
    "SO2"  : ["mean"],
    "NO2"  : ["mean"],
    "CO"   : ["mean"],
    "O3"   : ["mean"]
}).round(2)
```

- Berdasarkan tampilan tabel sebelumnya, dilakukan perhitungan rata-rata polutan di setiap kota.
- Perhitungan _mean_ polutan dari 12 kota dilakukan dengan menggunakan function **mean(axis=1)**.

```python
#Mean of Pollutants on Selected Stations
stations = airquality_df.groupby(by= 'station').agg({
    "PM2.5": ["mean"],
    "PM10" : ["mean"],
    "SO2"  : ["mean"],
    "NO2"  : ["mean"],
    "CO"   : ["mean"],
    "O3"   : ["mean"], 
}).mean(axis=1).round(2)
print(stations)
```

- Selanjutnya, mencari nilai rata-rata polutan dari 12 kota di Beijing berdasarkan jenis polutannya. 
- Perhitungan menggunakan fungsi **mean()** setelah function agg() dengan **round(2)** untuk membulatkan hasil rata-rata

```python
#Mean of Selected Pollutants
pollutants = airquality_df.groupby(by= 'station').agg({
    "PM2.5": ["mean"],
    "PM10" : ["mean"],
    "SO2"  : ["mean"],
    "NO2"  : ["mean"],
    "CO"   : ["mean"],
    "O3"   : ["mean"]
}).mean().round(2)
print(pollutants)
```

## Visualization & Explanatory Analysis

- Bagaimana tingkat polutan pada 12 kota di Beijing tahun 2013-2017?
- Apa tingkat polutan yang paling dominan antara polutan PM2.5, PM10, SO2, NO2, CO, dan O3 pada 12 kota di Beijing tahun 2013-2017?

### Pertanyaan 1:

- Visualisasi data ditampilkan dengan **Bar Chart** untuk menampilkan nilai rata-rata polutan di setiap 12 kota di Beijing tahun 2013-2017.

```python
#DataFrame
mean_of_pollutants = airquality_df.groupby(by= 'station').agg({
    "PM2.5": ["mean"],
    "PM10" : ["mean"],
    "SO2"  : ["mean"],
    "NO2"  : ["mean"],
    "CO"   : ["mean"],
    "O3"   : ["mean"], 
}).mean(axis=1).round(2)

#Show the figure and its size
plt.figure(figsize=(12, 7))

#Bar Chart
mean_of_pollutants.plot(kind='barh',
                    color='#ad074a',
                    label='Mean',
                    fontsize=10,
                    width=0.8)

#Chart Title
plt.title('Tingkat Polutan di Beijing Tahun 2013-2017',
          fontsize=14, fontweight='bold')

#X-Axes Label
plt.xlabel('Rata-Rata Polutan',
           fontsize=12, fontweight='bold')

#Y-Axes Label
plt.ylabel('Kota',
           fontsize=12, fontweight='bold')

#Dotted Line
plt.grid(axis='x',
         linestyle='--',
         alpha=0.5) 

#Add Legenda
plt.legend()

#Adjust The Layout
plt.tight_layout()

#Show the Plot
plt.show()
```

### Pertanyaan 2:

- Visualisasi data ditampilkan dengan **Bar Chart** untuk menampilkan nilai rata-rata setiap jenis polutan di Beijing tahun 2013-2017.

```python
#DataFrame
pollutants = airquality_df.groupby(by= 'station').agg({
    "PM2.5": ["mean"],
    "PM10" : ["mean"],
    "SO2"  : ["mean"],
    "NO2"  : ["mean"],
    "CO"   : ["mean"],
    "O3"   : ["mean"]
}).mean().round(2)

#Show the figure and its size
plt.figure(figsize=(10, 7))

#Bar Chart
pollutants.plot(kind='bar',
                color='#ad074a',
                label='Mean',
                fontsize=10,
                width=0.7)

#Chart Title
plt.title('Mayoritas Polutan di Beijing Tahun 2013-2017',
          fontsize=14, fontweight='bold')

#X-Axes Label
plt.xlabel('Jenis Polutan',
            fontsize=12, fontweight='bold')

#Y-Axes Label
plt.ylabel('Rata-Rata Polutan',
            fontsize=12, fontweight='bold')

#Rotate the X-Axes Label
plt.xticks(rotation=0)

#Dotted Line
plt.grid(axis='y', linestyle='--', alpha=0.5) 

#Add Legenda
plt.legend()

#Adjust The Layout
plt.tight_layout()

#Show the Plot
plt.show()
```

## Conclusion

- **Conclution pertanyaan 1**

    Rata-rata tingkat polutan pada 12 kota di Beijing tahun 2013-2017 tidak terdapat perbedaan yang signifikan. Selain itu, dapat diketahui bahwa kota dengan rata-rata polutan terbesar yaitu Kota Wanshouxigong dengan rata-rata polutan sebesar 283.33 m/m3.

- **Conclution pertanyaan 2**

    Jenis polutan yang paling dominan pada 12 kota di Beijing tahun 2013-2017 yaitu CO (Carbon Monoxide) dengan rata-rata polutan sebesar 1235.68 m/m3. Kandungan CO pada udara di Beijing memiliki perbedaan yang signifikan dengan kandungan polutan lainnya.
